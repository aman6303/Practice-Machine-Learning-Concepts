```python
import pandas as pd 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import precision_score
```

```python
data = load_iris()
x_train,x_test,y_train,y_test = train_test_split(data.data,data.target, random_state = 2)
```

```python
algo1 = LogisticRegression()
algo2 = DecisionTreeClassifier()

algo1.fit(x_train,y_train)
algo2.fit(x_train,y_train)

y_pred1 = algo1.predict(x_test)
y_pred2 = algo2.predict(x_test)

print('Logistic Regression precision score :\n',precision_score(y_test,y_pred1,average = None))
print('Decision tree precision score :\n',precision_score(y_test,y_pred2,average = None))
```

```python
print('Logistic Regression overall macro precision score :\n',precision_score(y_test,y_pred1,average = 'macro'))
print('Decision tree overall macro precision score :\n',precision_score(y_test,y_pred2,average = 'macro'))
```

```python
print('Logistic Regression overall weighted precision score :\n',precision_score(y_test,y_pred1,average = 'weighted'))
print('Decision tree overall weighted precision score :\n',precision_score(y_test,y_pred2,average = 'weighted'))
```

```python
# precision = true prediction / total prediction of that class
```
