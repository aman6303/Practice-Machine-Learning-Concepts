```python
import pandas as pd 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
```

```python
data = load_iris()
X = data.data
df = pd.DataFrame(X,columns = data.feature_names)
```

```python
df.head()
```

```python
df['target'] = data.target
```

```python
df.sample(5)
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('target',axis = 1),df.target,random_state = 2)
```

```python
algo1 = LogisticRegression()
algo2 = DecisionTreeClassifier()

algo1.fit(x_train,y_train)
algo2.fit(x_train,y_train)

y_pred1 = algo1.predict(x_test)
y_pred2 = algo2.predict(x_test)

print('Logistic Regression accuracy :',accuracy_score(y_test,y_pred1))
print('Decision tree accuracy :',accuracy_score(y_test,y_pred2))
```

```python
# we can also find score as
algo1.score(x_test,y_test),algo2.score(x_test,y_test)
```
