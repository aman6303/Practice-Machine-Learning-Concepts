```python
# f1 score = harmonic mean of recall and precison (HM favour/tilted lowers value)
```

```python
import pandas as pd 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import f1_score
```

```python
data = load_iris()
x_train,x_test,y_train,y_test = train_test_split(data.data,data.target, random_state = 2)
```

```python
algo1 = LogisticRegression()
algo2 = DecisionTreeClassifier()

algo1.fit(x_train,y_train)
algo2.fit(x_train,y_train)

y_pred1 = algo1.predict(x_test)
y_pred2 = algo2.predict(x_test)

print('Logistic Regression f1 score :\n',f1_score(y_test,y_pred1,average = None))
print('Decision tree f1 score :\n',f1_score(y_test,y_pred2,average = None))
```

```python
print('Logistic Regression overall macro f1 score :\n',f1_score(y_test,y_pred1,average = 'macro'))
print('Decision tree overall macro f1 score :\n',f1_score(y_test,y_pred2,average = 'macro'))
```

```python
print('Logistic Regression overall weighted f1 score :\n',f1_score(y_test,y_pred1,average = 'weighted'))
print('Decision tree overall weighted f1 score :\n',f1_score(y_test,y_pred2,average = 'weighted'))
```
