```python
from sklearn.datasets import make_classification,load_iris
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn import model_selection
from mlxtend.plotting import plot_decision_regions
```

```python
x, y = make_classification(
    n_samples=100,       # Number of rows
    n_features=2,        # Total number of columns
    n_classes=2,          # Binary classification (0 or 1)
    n_informative=2,    # Set this to match n_features
    n_redundant=0,      # Set this to 0 (Default was 2)
    n_repeated=0,
    random_state=42       # Ensures the same random numbers are generated each time
)
```

```python
plt.scatter(x[:,0],x[:,1],c = y)
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y)
```

```python
clf = SVC(kernel = 'rbf')
clf.fit(x_train,y_train)
```

```python
clf.intercept_
```

```python
y_pred = clf.predict(x_test)
```

```python
plt.scatter(x[:,0],x[:,1],c = y)
plot_decision_regions(x,y,clf = clf)
plt.show()
```
