```python
from sklearn.datasets import load_diabetes

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd = SGDRegressor(learning_rate='constant',eta0=0.1)
```

```python
import random
batch_size = 35

for i in range(100):  # we are doing 100 iteration by randomly selecting 35 rows at a time
    
    idx = random.sample(range(X_train.shape[0]),batch_size)
    sgd.partial_fit(X_train[idx],y_train[idx])   # we are training batch by batch using incremental learning(online learning)
```

```python
idx = random.sample(range(X_train.shape[0]),batch_size)
X_train[idx]
```

```python
sgd.coef_
```

```python
sgd.intercept_
```

```python
y_pred = sgd.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
# lets make batch size = no of rows --- batch gradient descent

batch_size = 100

for i in range(1):  # we are doing 1 iteration by  selecting all rows at a time
    
    idx = random.sample(range(X_train.shape[0]),batch_size)
    sgd.partial_fit(X_train[idx],y_train[idx]) 
```

```python
print(sgd.intercept_)
print(sgd.coef_)
y_pred = sgd.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# lets make batch size as 1 p-- stochastic gradient descent

batch_size = 1

for i in range(1):  # we are doing 1 iteration by  selecting all rows at a time
    
    idx = random.sample(range(X_train.shape[0]),batch_size)
    sgd.partial_fit(X_train[idx],y_train[idx]) 
```

```python
print(sgd.intercept_)
print(sgd.coef_)
y_pred = sgd.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# comparing this with inbuild stochastic

sgd = SGDRegressor(learning_rate='constant')
sgd.fit(X_train,y_train) 
print(sgd.intercept_)
print(sgd.coef_)
y_pred = sgd.predict(X_test)
r2_score(y_test,y_pred)
```
