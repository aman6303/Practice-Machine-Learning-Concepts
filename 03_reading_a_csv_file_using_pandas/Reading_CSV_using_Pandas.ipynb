# 11. uisng converters

```python
import pandas as pd
from io import StringIO

# Create a small CSV text for demo
csv_data = """ID,Name,Salary,Joining_Date
1,John,45000,2020-01-10
2,Alice,55000,2021-03-15
3,Bob,60000,2022-05-22
4,David,50000,2020-12-01
"""

# Use StringIO to simulate reading from a file
data = StringIO(csv_data)

# Define converter functions
def convert_salary(x):
    return float(x) / 1000   # convert salary to 'thousands'

def convert_date(x):
    return pd.to_datetime(x).strftime("%b-%Y")  # format date

# Read CSV using converters
df = pd.read_csv(
    data,
    converters={
        "Salary": convert_salary,      # apply to Salary column
        "Joining_Date": convert_date   # apply to Joining_Date column
    }
)

print(df)
```
