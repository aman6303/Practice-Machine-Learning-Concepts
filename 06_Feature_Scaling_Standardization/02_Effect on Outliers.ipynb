```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# importing dataset
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```python
# lets impute some outliers in the data

df = pd.concat([df,pd.DataFrame({'Age':[5,90,100],'EstimatedSalary':[1000,200000,300000]})],ignore_index=True)
df.shape
```

```python
x = df.iloc[:,0:2]
x.head()
```

```python
y= df.iloc[:,2]
y.head()
```

```python
# splitting the dataset
from sklearn import model_selection

x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```
