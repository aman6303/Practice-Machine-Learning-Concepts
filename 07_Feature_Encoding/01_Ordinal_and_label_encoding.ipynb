```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
# # we perform ordinal encoding on ordical data for input values
```

```python
df = pd.read_csv('customers.csv')
df.head()
```

```python
# we will be performing Ordinal encoding on review and education and lbel encoding on target variable
```

```python
df_new = df.iloc[:,2:]
df_new.head()
```

```python
# splitting the data

from sklearn import model_selection
x_train, x_test , y_train, y_test  = model_selection.train_test_split(df_new.iloc[:,0:2],df_new
    .iloc[:,2])
```

```python
x_train
```

```python
# lets perform ordinal encoding on input data

oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
oe.fit(x_train)
```

```python
oe.categories_
```

```python
# tranforming the values

x_train = oe.transform(x_train)
x_train
```

```python
## applying label encoder on target varibale
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()  # we not need to provide categories in case of label encoder
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train
```

```python
# transforming the value
y_train = le.transfrom(y_train)
y_train
```
