## 2. One hot encoding using sklean

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# creating object without parse and dummy trap and dtype
# wew should perform it after splitting but in this case i am doing it direct

ohe = OneHotEncoder()

x_train_new = ohe.fit_transform(df.iloc[:,2:4]).toarray()  # we manually need to convert it
x_train_new
```

```python
# we need to add this in initial df 

x_train = np.hstack((df[['brand','km_driven','selling_price']].values,x_train_new)) #combining the values and take tuples
x_train
```

```python
# creating object with parse and dummy trap and dtype
ohe = OneHotEncoder(drop='first',sparse_output = False, dtype = int)

x_train_new = ohe.fit_transform(df.iloc[:,2:4])   # we manually need to convert it
x_train_new
```

```python
x_train = np.hstack((df[['brand','km_driven','selling_price']].values,x_train_new)) #combining the values and take tuples
x_train
```

```python
counts = df.brand.value_counts()
```

```python
# aplying on brand categories

ohe = OneHotEncoder(drop='first',sparse_output = False, dtype = int)

repl = counts[counts<=9].index  # getting index 
df['brand'] = df['brand'].replace(repl,'uncommon')
x_train_new = ohe.fit_transform(df[['brand']]) 
x_train_new
```
