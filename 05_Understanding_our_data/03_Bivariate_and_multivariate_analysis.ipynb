```python
import pandas as pd
import seaborn as sns
```

## 1.Scatterplot (num vs num)

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(x=df['total_bill'],y=df['tip'])
```

```python
sns.scatterplot(x=df['total_bill'],y=df['tip'], hue = df['sex'],size=df['size'])
```

```python
## 2. Barplot(num vs categorical)
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.barplot(x=df.pclass, y=df.age)
```

```python
# black line show confidence interval
```

```python
sns.barplot(x=df.pclass, y=df.fare, hue = df.sex)
```

```python
sns.barplot(x=df.pclass, y=df.age, hue = df.sex)
```

```python
## 3.Box plot (num - cat)
```

```python
sns.boxplot(x=df.sex, y=df.age)
```

```python
sns.boxplot(x=df.sex, y=df.age, hue=df.pclass)
```

```python
## 4.distplot()
```

```python
sns.distplot(df.age)
```

```python
# age vs survived

sns.distplot(df[df.alive=='yes'].age, hist=False)
sns.distplot(df[df.alive=='no'].age, hist=False)
```

```python
## 5.heatmap (cat-cat) (draw crosstab or pivot_table)
```

```python
#creating a cross tab
pd.crosstab(df.pclass,df.alive)
```

```python
sns.heatmap(pd.crosstab(df.pclass,df.alive))
```

```python
df = sns.load_dataset('flights')
df.head()
```

```python
# creating a pivot_table

pd.pivot_table(data = df, index = 'month', values = 'passengers', columns = 'year')
```

```python
sns.heatmap(pd.pivot_table(data = df, index = 'month', values = 'passengers', columns = 'year'))
```

```python
## 6. Barplot with percentage values
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.groupby('sex').mean('survived').reset_index()
```

```python
df.groupby('sex')['survived'].mean().reset_index().plot(kind='bar')
```

```python
## 7. clustermap - heatmap with relationship and dendogram
```

```python
# on titanic dataset
df = sns.load_dataset('titanic')
sns.clustermap(pd.crosstab(df.pclass,df.alive))
```

```python
df = sns.load_dataset('flights')
```

```python
sns.clustermap(pd.pivot_table(data = df, index = 'month', values = 'passengers', columns = 'year'))
```
