```python
from sklearn.experimental import enable_iterative_imputer  # to run iterativeimputer we need this 
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('50_Startups.csv')
```

```python
new_df = df.head().copy()
```

```python
new_df
```

```python
new_df = new_df.iloc[:,0:3].copy()
```

```python
new_df
```

```python
# introducing missing values
new_df.iloc[1,0] = np.nan
new_df.iloc[3,1] = np.nan
new_df.iloc[-1,-1] = np.nan
```

```python
new_df
```

```python
# now we will be filling those values and compare with original one

# Configure imputation engine using advanced non-linear learners
mice_imputer = IterativeImputer(
    estimator=RandomForestRegressor(),   # we can use any estimator
    max_iter=10,
    initial_strategy="mean",
    imputation_order="ascending"
)

df_imputed = pd.DataFrame(
    mice_imputer.fit_transform(new_df),
    columns=new_df.columns
)
new_df
```
